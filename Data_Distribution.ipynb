```python
import numpy as np

a=np.random.uniform(0,5,250)
print(a)
```

```python
#using matplotlib to visualize our dataset using a histogram
import matplotlib.pyplot as plt

plt.hist(a,5)
plt.show()
```

```python
#Create an array with 100000 random numbers, and display them using a histogram with 100 bars

b=np.random.uniform(0,50,10000)  #(start,end,total no of values to generate)
plt.hist(b,100) #the hist(array_values,no of bars to represent)
plt.show()
```
