```python
import pandas as pd
df=pd.read_csv("data1.csv")

#all values for a decision tree must be in numerals so we must assign an integer value for all non integer values
print(df)

```

```python
#we will use the map() of pd to do that
d={'UK':0,'USA':1,'N':2}
df["Nationality"]=df["Nationality"].map(d)
print(df["Nationality"])
```

```python
#now we will do the same thing for GO column
d={'NO':0,'YES':1}
df['Go']=df['Go'].map(d)
print(df['Go'])
```

```python
#the next step is to separate the features and the target column, features are all those column using which we are 
#going to use to make the prediction
feature=['Age','Experience','Rank','Nationality']
X=df[feature]  #features
y=df['Go']     #Target
```

```python
#now time for the actual decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

dtree=DecisionTreeClassifier() #this will create the tree object where we will pass the variables
dtree=dtree.fit(X,y)

tree.plot_tree(dtree, feature_names=feature)
```

```python
print(dtree.predict([[40, 10, 7, 1]]))
print(dtree.predict([[40, 10, 6, 1]]))
```
