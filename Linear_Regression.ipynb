```python
#so regression is used to find the relationship between variables and this relationship is used in statistical 
#modeling and machine learning to predict future events or outcomes
#linear regression uses the relationship between data points to draw a straight line and using that line we can
#predict the future or future values

import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6] #ages
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] #speed at toolbooth

plt.scatter(x,y)
plt.show()
```

```python
#we can use scipy to draw a line of linear regression from the scatter plot
import scipy as sc
#we will use the same x and y values from the top

#i don't understand this part
slope,intercept,r,p,std_err = sc.stats.linregress(x,y)

def myfunc(x):
    return slope * x + intercept

mymodel=list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show() 
```
