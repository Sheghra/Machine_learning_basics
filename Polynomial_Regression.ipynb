```python
#the data is of cars passing at various speeds through a toolbooth at different times of the day
import matplotlib.pyplot as plt
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22] #time of day
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100] #speed of the car

plt.scatter(x,y)
plt.show()
```

```python
#we will use the same values
import numpy

mymodel=numpy.poly1d(numpy.polyfit(x,y,3))

print(numpy.polyfit(x,y,3))
#print(numpy.poly1d(numpy.polyfit(x,y,3)))


myline=numpy.linspace(1,22,100)

plt.scatter(x,y)
plt.plot(myline,mymodel(myline))
plt.show()
```

```python
#R-Squared using sklearn
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel=numpy.poly1d(numpy.polyfit(x,y,3))  #this is the model which will predict y values from x-values

print(r2_score(y,mymodel(x)))
```

```python
#now we can predict future values since we checked our r2_score of our model
print(mymodel(17))
```
