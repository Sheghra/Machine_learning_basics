```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)

x=np.random.normal(3,1,100)
y=np.random.normal(150,40,100)/x

plt.scatter(x,y)
plt.show()
```

```python
#split the dataset into two sets 80 for training and the rest 20 for testing
train_x=x[:80]  #start from 0 go till 80-1=79 80 values out of 100 into the training set
train_y=y[:80]

test_x=x[80:] #start from 80 and then go on till n-1 that is 20 into the testing set
test_y=y[80:]

#we have successfully separated the datasets into two, one for training the model and the other for testing the
#accuraccy of the trained model
```

```python
#displaying the training set
plt.scatter(train_x,train_y)
plt.show()
```

```python
#displaying the testing set
plt.scatter(test_x,test_y)
plt.show()
```

```python
#then fit the dataset, look at the plotted out and figure which regression would be best suited for it, 
#let's use polynomial regression for it after analysing the data

import numpy as np 
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
np.random.seed(2)

x=np.random.normal(3,1,100)
y=np.random.normal(150,40,100)/x

train_x=x[:80]  #start from 0 go till 80-1=79 80 values out of 100 into the training set
train_y=y[:80]

test_x=x[80:] #start from 80 and then go on till n-1 that is 20 into the testing set
test_y=y[80:]

mymodel=np.poly1d(np.polyfit(train_x,train_y,4))

myline=np.linspace(0,6,100) #these values can be decided by looking at the data in the scatter plots

r2=r2_score(train_y,mymodel(train_x))
print(r2)

plt.scatter(train_x,train_y)
plt.plot(myline,mymodel(myline))
plt.show()

#we have tested our model on the training data and the r2 score is 0.798 which shows a good relationship and 
#our model is ok, so now, let's test it with the testing data and see how well it performs
```

```python
print(r2_score(test_y,mymodel(test_x))) 
#the score is well enough and close to 1 so it shows that the model fits the training set well and we can be confident
#enough to use this model for future predictions
```

```python
#once our model has been tested we can use it to predict future values,
#let's see how much will a customer spending 6 minutes in a shop will spend
print(mymodel(5))
```
